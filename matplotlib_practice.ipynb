```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot
plt.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green")
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create lables for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare(&)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="purple", marker="8")
ax.legend()
# OR
fig = plt.figure()
ax = plt.axes()
ax.plot(x_axis, y_axis, label="Boston", color="orange", marker="h")

# Create lables for the x and y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# # Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# # Create a title
ax.set_title("PyBer Fare by Month")

# # Add a grid
ax.grid()

# # Add the legend
ax.legend()
```

```python
# Create plot for bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create plot for bar chart
plt.bar(x_axis, y_axis, color='green', label='Boston')
# Label axes
plt.xlabel("Date")
plt.ylabel('Fare($)')
# Create title
plt.title('Pyber Fare by Month')
# Add legend to chart
plt.legend()
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()

# Add title and axes labels
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
```

```python
# Create plot using ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)

# OR
fig = plt.figure()
ax = plt.axes()
ax.bar(x_axis, y_axis)
```

```python
# Create bar chart with ax.plt() (object-oriented interface)
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

# Switch/invert y axis labels with Jan at top
ax.invert_yaxis()
# Label axes and add title
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create Legend
ax.legend()
```

```python
# Create scatter plot using plt.plot()
# plt.plot(x_axis, y_axis, "o")
# Create scatter plot using plt.scatter()
plt.scatter(y_axis, x_axis, color="red", label="Chicago")

# Invert date axis
plt.gca().invert_yaxis()
# Label axes, add title
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Add limit to x axis
plt.xlim(0, 45)
# Add legen
plt.legend()
```

```python
# Create scatter plot and use for loop to increase size of data markers
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create scatter plot and use python list comprehension to increase size of data points
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
# Create scatter plot using object oriented method
fix, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s= [i*5 for i in y_axis], color='skyblue', alpha=.8, edgecolor='k', linewidth=2, label='Boston')

# Invert y data labels
ax.invert_yaxis()
# Label axes, add title, legend, x lim
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_xlim(0,50)
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', "tomato", 'mistyrose']
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct="%.1f%%")

plt.show()
```
